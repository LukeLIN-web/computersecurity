```python
import matplotlib.pyplot as plt

# Initialize empty lists to store the data
x_data = []
y_data = []

with open('study.400', 'r') as file:
    for line in file.readlines()[:256]:
    # for line in file.readlines()[7425:7680]: # 13
        values = line.strip().split()
        x_data.append(float(values[2]))  # Column 3
        y_data.append(float(values[6]))  # Column 6
        print(values)

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, marker='o', color='b')
plt.xlabel('n[i]')
plt.ylabel('OpenSSL AES timings (cycles)')
plt.title('How n[i] affects OpenSSL AES timings ')
plt.legend()
plt.grid(True)

# Show the plot or save it to a file
plt.show()
```
